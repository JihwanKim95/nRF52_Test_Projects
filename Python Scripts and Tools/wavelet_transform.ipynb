```python
import pandas as pd 
import numpy as np 
import scaleogram as scg 
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('sample.csv')
df = df.iloc[1:201,0:]
df
```

```python
class_labels = df.columns[1:]
X = df.iloc[:, 0:].T.values
print(X.shape)
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
x = X[0]
```

```python
x1 = X[0]
x2 = X[1]
x3 = X[2]
x1
```

```python
win_size = 5 
sval = 0.002
wavelet='cmor0.7-1.5'

x1 = np.asarray(x1, dtype=np.float32)
x2 = np.asarray(x2, dtype=np.float32)
x3 = np.asarray(x3, dtype=np.float32)

time = np.arange(0, x.size) # * 1/45 
scales = np.arange(1, 64, dtype=np.int32)

ax1 = scg.cws(time, signal=x1, scales=scales, wavelet=wavelet, cmap='rainbow', 
             cbar=None, yaxis='frequency', ylabel=None, xlabel=None, yscale=None, coi=False)

ax2 = scg.cws(time, signal=x2, scales=scales, wavelet=wavelet, cmap='rainbow', 
             cbar=None, yaxis='frequency', ylabel=None, xlabel=None, yscale=None, coi=False)

ax3 = scg.cws(time, signal=x3, scales=scales, wavelet=wavelet, cmap='rainbow', 
             cbar=None, yaxis='frequency', ylabel=None, xlabel=None, yscale=None, coi=False)

# print(ax1)
# print(ax2)
# print(ax3)
```
